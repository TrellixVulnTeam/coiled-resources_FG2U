```python
import coiled
import dask.bag as db
from dask.distributed import Client
import json
```

```python
cluster = coiled.Cluster(
    name="github-json",
    software="matthew-powers/my_env", 
    n_workers=5,
)
```

```python
client = Client(cluster)
```

```python
rows = db.read_text(
    "s3://coiled-datasets/github-archive/github-archive-2015-json/*.json.gz",
    compression="gzip"
).map(json.loads)
```

```python
rows.filter(lambda d: d["user"] == "MrPowers").compute()
```

```python
rows.take(3)
```

```python
rows.map(lambda d: d["repo"]).frequencies().take(20)
```

```python
rows.map(lambda d: d["repo"]).frequencies(sort=True).take(20)
```

```python
rows.filter(lambda d: d["user"] == "MrPowers").map(lambda d: d["repo"]).frequencies(sort=True).compute()
```
