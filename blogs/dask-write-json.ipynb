```python
import dask.dataframe as dd
import pandas as pd
```

```python
df = pd.DataFrame({"col1": ["a", "b", "c", "d"], "col2": [1, 2, 3, 4]})
```

```python
ddf = dd.from_pandas(df, npartitions=2)
```

```python
ddf.to_json("../tmp/first-json")
```

```python
ddf.to_json("../tmp/json2", orient="split")
```

```python
def my_json_filenamer(file_num):
    return f"part-{file_num:04}-data"
```

```python
ddf.to_json("../tmp/json3", name_function=my_json_filenamer)
```

```python
def my_csv_filenamer(file_num):
    return f"some-{file_num:04}-data.csv"
```

```python
ddf.to_csv("../tmp/csv3", name_function=my_csv_filenamer)
```

```python
import dask
```

```python
ddf = dask.datasets.timeseries(
    start="2015-01-01",
    end="2020-12-31",
    freq="1s",
    partition_freq="7d",
    seed=42,
)
```
